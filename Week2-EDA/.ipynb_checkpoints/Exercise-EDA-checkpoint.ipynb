# It seems as though in the years before 2000, they were counting in tens.
Make a new column that is called `adjusted_price`, that contains all prices, however, for 
every year before 2000 make it 10x the original MSRP.  
_Hint use `np.where()`_.

```python
# Make a new column that is called `adjusted_price`, that contains all prices, however, 
# for every year before 2000 make it 10x the original MSRP.

before = df['Year'] < 2000

df['adjust_price']= np.where(before, df['MSRP']*10, df['MSRP']*1)

df[before].head()
```

# Replot the new adjusted price.  
* Make the y-axis start at 0 and go up to 100,000

```python
# Replot the new adjusted price and make the y-axis start at 0 and go up to 100,000

replot = sns.lineplot(data=df, x='Year', y='adjust_price');

replot.set_yticks(ticks=[0,100000]);

replot.set_title('Adjusted price and Year');
```

# Remove any statisical outliers from Engine HP and highway MPG and plot the result.
* Statisical outliers meaning values that are further than 3 standard deviations from the mean 
* Create a new columns for z-scores for each 'Engine HP' and 'highway MPG' named 'Engine HP_zscore' and 'highway MPG_zscore'
* Calculate the z-scores for each of our target columns.
* Make sure you set the z-scores to be their absolute values. 
* Create condition masks for when either of those absolute values are greater than 3.

* Create a new dataframe that is a copy of our dataframe using df.copy()

* Using our new dataframe
    * Filter out all Engine HP Z-Scores that are greater than 3
    * Filter out all Highway MPG z-scores that are greater than 3.
    * Make the same scatterplot plotting the relationship of Engine HP and Highway MPG as before but with the this dataframe.

```python
# Remove any statisical outliers from Engine HP and highway MPG

y = ["Engine HP", "highway MPG"]

for x in y:
    new_name = x + "_zscore"
    
    #Calculate the z-scores for each of our target columns.
    df[new_name] = stats.stats.zscore(df[x])
    
    #Make sure you set the z-scores to be their absolute values.
    df[new_name] = abs(df[new_name])
    
    
# Create a new dataframe that is a copy of our dataframe using df.copy()    
new_df = df.copy()

condition1 = df['Engine HP_zscore'] > 3
condition2 = df['highway MPG_zscore'] > 3

########
new_df.plot(kind = "scatter", y ="Engine HP" , x = "highway MPG" )

```
